#Ask the user for a number. Depending on whether the number is even or odd, print out an appropriate message to the user.  Hint: how does an even / odd number react differently when divided by 2?

```python
# A program to check a number is odd or even
def oddEven(num):
  if num % 2 == 0:
    print(f"{num} is even")
  else:
    return f"{num} is odd"

while True:
  num = int(input("Please enter a number: "))
  print(oddEven(num))
  cont = input("Do you want to check a new number? (y/n): ")
  if cont == "n":
    print("Bye.")
    break
```

```python
# A program to check a number is odd or even
num = int(input('Enter a number: '))

if num % 4 == 0:
    print('The number you input is a multiple of 4')
    
elif num % 2 == 0:
    print('The number you input is an even')
else:
    print('The number you input is an odd')

# Ask the user for two numbers
num = int(input('Enter the first number: '))
check = int(input('Enter the second number: '))

if num % check == 0:
    print('The first number could be divided evenly by the second number')
else:
    print('The first number could not be divided evenly by the second number')


```
